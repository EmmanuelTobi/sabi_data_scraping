```python
import os
import pandas as pd
from pathlib import Path

#os.system('pip3 install pandas')

import requests
import json

from bs4 import BeautifulSoup
```

```python
#create a easy functions to automate our flow

def getSoupDataFromLink(url):
    html = requests.get(url).text
    soup = BeautifulSoup(requests.get(url).content, "html.parser")
    
    return soup

def getWordsFromSoup(data):

    extracted_words = []
    
    mainBody = data.find('body').find_all('h3')
    newsBody = data.find('body').find_all('p')

    newsBody_list = [x.text.strip() for x in newsBody]
    headlines_list = [x.text.strip() for x in mainBody]
    
    extracted_words.extend(newsBody_list)
    extracted_words.extend(headlines_list)

    print()

    return extracted_words



def getMoreLinksFromArticles(data, link_check):
    moreLinks = []
    
    link_lists = [x['href'] for x in data.find_all('a', href=True)]
    for x in link_lists:
        if x.find(link_check) != -1:
            moreLinks.append(x)
                
    return moreLinks, len(moreLinks)




def turnListOfWordsToDF (data, counting_len, stop_len, file_name, link):    
    df = pd.DataFrame(data)
    
    sorted(data)

    if(counting_len == stop_len):
        
        filepath = Path(file_name)  
        filepath.parent.mkdir(parents=True, exist_ok=True)
        
        df['lang'] = ["hausa"] * df[df.columns[0]].count()
        df['links'] = [link] * df[df.columns[0]].count()
        
        df.to_csv (filepath, index = None, header=True)

    print(df)



def extractFromSetLinkList(links_to_check):

    extractedListOfWords = []

    soup_data = getSoupDataFromLink(links_to_check)
    more_links_gotten = getMoreLinksFromArticles(soup_data, "https://www.bbc.com/hausa/articles")[0]

    a = 0

    for datax in more_links_gotten:
        a = a + 1
        filename = 'sabifolder/newsData/extractedNewsData00' + str(a) + ".csv"

        #getMoreLinksFromArticles(getSoupDataFromLink(datax), "https://www.bbc.com/hausa/articles")

        soup = getSoupDataFromLink(datax)
        words = getWordsFromSoup(soup)
        extractedListOfWords.extend(words)
        turnListOfWordsToDF(extractedListOfWords, a, len(more_links_gotten), filename, datax)


links_to_test = ["https://www.bbc.com/hausa/articles/c33v1pp0v5ko", 
                 "https://www.bbc.com/hausa/articles/c78d7499w94o",
                 "https://www.bbc.com/hausa/articles/cwyl0r1v403o",
                 "https://www.bbc.com/hausa/articles/cn03yl0nwz9o"]

single_to_test = ["https://www.bbc.com/yoruba/articles/c99vxzpj722o"]



for l in links_to_test:
    extractFromSetLinkList(l)
    
```
